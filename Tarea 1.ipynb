```python
import matplotlib.pyplot as plt

anios = [
    1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999,
    2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
    2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,
    2020, 2021, 2022, 2023, 2024
]

energia = [
    106495, 107243, 107963, 108860, 110233, 112519, 115579, 116736, 117456, 119413,
    122288, 123390, 125796, 129561, 135359, 139327, 142782, 146844, 148312, 146047,
    152462, 155655, 157617, 160030, 161541, 162544, 164326, 167697, 171839, 173646,
    167913, 175926, 178901, 181912, 186383
]

plt.figure(figsize=(12,6))
plt.plot(anios, energia, marker='o', linestyle='-', color='b', label="Energía consumida (TWh)")

plt.title("Consumo de Energía (1990 - 2024)", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Energía consumida (TWh)", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend()
plt.tight_layout()

plt.show()
```

```python
import numpy as np

años = np.array([
    1990, 1991, 1992, 1993, 1994, 1995, 1996, 1997, 1998, 1999,
    2000, 2001, 2002, 2003, 2004, 2005, 2006, 2007, 2008, 2009,
    2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019,
    2020, 2021, 2022, 2023, 2024
])

energia = np.array([
    106495, 107243, 107963, 108860, 110233, 112519, 115579, 116736, 117456, 119413,
    122288, 123390, 125796, 129561, 135359, 139327, 142782, 146844, 148312, 146047,
    152462, 155655, 157617, 160030, 161541, 162544, 164326, 167697, 171839, 173646,
    167913, 175926, 178901, 181912, 186383
])


grado = 2  
coef = np.polyfit(años, energia, grado)
modelo = np.poly1d(coef)


crecimiento_promedio = (energia[-1] - energia[0]) / (años[-1] - años[0])
print(f"Ritmo de crecimiento promedio anual: {crecimiento_promedio:.2f} TWh por año")


años_futuro = np.arange(2025, 2051)
energia_futuro = modelo(años_futuro)


plt.figure(figsize=(12,6))
plt.scatter(años, energia, color="blue", label="Datos históricos")
plt.plot(años, modelo(anios), "r--", label="Modelo ajustado")
plt.title("Consumo de Energía y Modelo de Crecimiento", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Energía consumida (TWh)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()


plt.figure(figsize=(12,6))
plt.plot(años_futuro, energia_futuro, "g-o", label="Proyección futura")
plt.title("Proyección del Consumo de Energía (2025-2050)", fontsize=14)
plt.xlabel("Año", fontsize=12)
plt.ylabel("Energía consumida (TWh)", fontsize=12)
plt.legend()
plt.grid(True, linestyle="--", alpha=0.6)
plt.tight_layout()
plt.show()
```
